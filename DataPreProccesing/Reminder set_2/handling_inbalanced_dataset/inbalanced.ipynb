```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('kyphosis.csv')
```

```python
data.sample(5)
```

```python
x=np.array(data.drop(columns=['Kyphosis']))
```

```python
y=data.Kyphosis
```

```python
y.value_counts().plot(kind='bar')
```

```python
from imblearn.under_sampling import RandomUnderSampler
under_sampler=RandomUnderSampler()
x_under,y_under=under_sample.fit_resample(x,y)
```

```python
y_under.value_counts().plot(kind='bar')
```

```python
from imblearn.over_sampling import RandomOverSampler
over_sampler=RandomOverSampler()
x_over,y_over=over_sampler.fit_resample(x,y)
```

```python
y_over.value_counts().plot(kind='bar')
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_smot,y_smot=smote.fit_resample(x,y)
```

```python
y_smot.value_counts().plot(kind='bar')
```
