```python
def add_company(input):
    if input=='samsung' or input=='Razer' or input=='Mwdiacom' or input=='Microsoft' or input=='Vero' or input=='Xiaomi' or input=='Google' or input=='Fujitsu' or input=='Chuwi' or input=='LG' or input=='Huawei':
        return 'Other'
    else:
        return input
data['Company']=data['Company'].apply(add_company)
```

```python
data['ScreenResolution'].value_counts()
```

```python
data['TouchScreen']=data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['IPS']=data['ScreenResolution'].apply(lambda x: 1 if 'IPS' in x else 0)
```

```python
data.head(2)
```

```python
data['Cpu']=data['Cpu'].apply(lambda x: ' '.join(x.split()[0:3]))
```

```python
data.head(2)
```

```python
data['Cpu'].value_counts()
```

```python
def set_processor(name):
    if name=='Intel Core i7' or name=='Intel Core i5' or name=='Intel Core i3':
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Other'
data['Cpu']=data['Cpu'].apply(set_processor)
```

```python
data['Cpu'].value_counts()
```

```python
data.head()
```

```python
data['Gpu'].value_counts()
```

```python
data['Gpu_name']=data['Gpu'].apply(lambda x:' '.join(x.split()[0:1]) )
```

```python
data.head(2)
```

```python
data['Gpu_name'].value_counts()
```

```python
data=data[data['Gpu_name']!='ARM']
```

```python
data.shape
```

```python
data['OpSys'].value_counts()
```

```python
def os(input):
    if input=='Windows 10' or input=='Windows 7' or input=='Windows 10 S':
        return 'Windows'
    elif input=='macOS' or input=='Mac OS X':
        return 'Mac'
    elif input=='Linux':
        return 'Linux'
    else:
        return 'Others'
data['OpSys']=data['OpSys'].apply(os)
```

```python
data.head(5)
```

```python
data['OpSys'].value_counts()
```

```python
data.head(2)
```

```python
data=data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Gpu'])
```

```python
data.head()
```

```python
data=pd.get_dummies(data)
```

```python
data.head()
```

```python
data.shape
```

```python
x=data.drop(['Price_euros'],axis=1)
y=data.Price_euros
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
print(len(xtrain))
print(len(xtest))
```

```python
acc_list=[]
    
def model_accuracy(model):
    model.fit(xtrain,ytrain)
    acc=model.score(xtest,ytest)
    acc_list.append(acc)
    print(model,acc)
acc_list
    
```

```python
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model_li=LinearRegression()
model_dc=DecisionTreeRegressor()
model_lss=Lasso()
model_rf=RandomForestRegressor()
model_list=[]
model_list.extend([model_li,model_dc,model_lss,model_rf])
```

```python
model_list
```

```python
for i in model_list:
    model_accuracy(i)
```
